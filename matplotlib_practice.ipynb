```python
#import dependencies
import matplotlib.pyplot as plt
```

```python
# LINE CHARTS
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Chart using object_oriented interface
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
# Create the plot with ax.plt()
#fig = plt.figure()
#ax = fig.add_subplot()
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with plt.show() * Warning - to be used once per session 
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Annotation and formatting with MatPlotLib library 
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="green", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# BAR CHARTS
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot and annotate
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the horizontal bar plot
# Create the plot.
#SWITCH AXIS using gca() otherwise data looks bad 
plt.barh(x_axis, y_axis, color="magenta", label='Boston')
plt.gca().invert_yaxis()
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Bar Charts and Object Oriented Approach 

# Create the vertical plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color="black", label ='New York')
```

```python
# Create the horizontal plot using ax.plt()
fig, ax = plt.subplots()
# Switch the axis 
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
ax.set_title("Pyber Fare by Month")
ax.set_ylabel('Date')
ax.set_xlabel('Fare($)')
ax.legend()
ax.grid()
```

```python
# CREATE SCATTER PLOTS & BUBBLE CHARTS

#Scatter plot with plt.plot(x,y, 'o') 'o' designates scatter
plt.plot(x_axis, y_axis, 'o')
```

```python
# Scatter plot with plt.scatter()
plt.scatter(y_axis, x_axis, color='red', label='Chicago')
plt.gca().invert_yaxis()
plt.legend()
plt.xlabel('Fare($)')
plt.ylabel('Date')
plt.title('PyBer Fare by Month')
plt.xlim(0,50)
```

```python
# BUBBLE CHART; new parameters of size = s 
#plt.scatter(x_axis, y_axis, s=y_axis)

# Make the bubles larger to see by multiplying at constant value 
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Refactor and use list comprehension instead of for loop 
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Scatter Plots with Object Oriented Interface 
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Bubble Chart with Object Oriented Interface 
fig, ax = plt.subplots()
plt.gca().invert_yaxis()
ax.scatter(y_axis, x_axis, color='skyblue', s=[i * 5 for i in y_axis], linewidth=2 , label='Boston')
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_xlim(0,50)
```

```python
# PIE CHARTS MATLAB & Object-oriented 
# PIE with MATLAB
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Add percentages to each month and emphasize the wedge with "explode"
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Change the colors with CSS colors in Matplotlib
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# Create Pie using Object-Oriented Interface 
fig, ax = plt.subplots()
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
ax.pie(y_axis, labels=x_axis, shadow=True, colors=colors, 
       explode=explode_values, counterclock=False, startangle=90, 
       radius=2, autopct='%.1f%%')





```
